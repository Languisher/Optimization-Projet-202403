```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
m = 18000
g = 9.8
f = 0.016
A = 8.16
C = 0.6
n = 0.86
v = np.linspace(0,60,10)
p = (m * g * f + C * A * (v**2) / 21.15) * v / (3600 * n)

r = 0.3
T = 800
b2 = 1.8
a2 = T / (3600 * r)
p2 = a2 * v + b2
print(a2,b2)
x = sp.Symbol('x')
equation1 = (m * g * f + C * A * (x**2) / 21.15) * x / (3600 * n)
equation2 = a2 * x + b2

solutions = sp.solve(sp.Eq(equation1, equation2), x)

v_int = solutions[0]
p_int = a2 * v_int + b2
print(v_int, p_int)  

plt.plot(v, p, color="b", marker='o',label='P_con', markerfacecolor='white')
plt.plot(v, p2 , color='red', label='P_rec') 
plt.scatter(v_int, p_int, color='g', label='CP',marker = 'o') 
plt.fill_between(v, p, p2,where=(p2 > p), color='palegreen', alpha=0.5)
plt.fill_between(v, p, p2,where=(p > p2), color='skyblue', alpha=0.5)


plt.title('La relation entre la puissance et la vitesse')
plt.grid(True)
plt.xlabel('Vitesse (km/h)')
plt.ylabel('Puissance (kw)')
plt.legend()
plt.show()
```
